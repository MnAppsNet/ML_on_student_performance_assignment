```python
# Drop the columns with high correlation
dataset_mat_filtered = dataset_mat.drop(["FathersEducation","MothersJob","WeeklyAlcohol","G2", "G3" ], axis=1)
dataset_por_filtered = dataset_por.drop(["FathersEducation","MothersJob","WeeklyAlcohol","G2", "G3" ], axis=1)
```

```python
import numpy as np

def map_values(value):
    if int(value) > 15:
        return 2
    elif int(value) > 10:
        return 1
    else:
        return 0

# Example dataset
#dataset = np.random.uniform(0, 20, size=100)  # Generate random dataset
column_name = 'G1'
# Apply mapping function to each value in the dataset
dataset_mat_filtered[column_name] = np.array([map_values(value) for value in dataset_mat_filtered[column_name]])
```

```python
X = dataset_mat_filtered.drop(["G1"], axis=1)
y = dataset_mat_filtered["G1"]

#Get 80% of data for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2, stratify=y)
y_train.value_counts()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report

models = [
    ['RandomForest',RandomForestClassifier(n_estimators=50, max_depth=20, criterion="gini", random_state=42)],
    ['LinearRegression',LinearRegression()],
    ['DecisionTree', DecisionTreeClassifier(criterion="entropy", max_depth=20)],
    ['SVM',SVC(degree=6,kernel = "poly")]
]

for m in models:
    print(m[0])
    m[1].fit(X_train, y_train)
    y_pred = [round(x) for x in m[1].predict(X_test)]
    print(classification_report(y_pred,y_test))
```

```python
rf = models[0][1]
plt.title("Random Forest - Feature Importance")
sns.barplot(y=X_train.columns,x=rf.feature_importances_, orient='h')
```

```python
lr = models[1][1]
plt.title("Linear Regression - Feature Importance")
sns.barplot(y=X_train.columns,x=lr.coef_, orient='h')
```

```python
# Calculate the confusion matrix
confusion = confusion_matrix(y_test, models[0][1].predict(X_test))

# Define class labels
class_labels = ['NotPass', 'Pass', 'HighPass']

# Create a DataFrame for the confusion matrix
confusion_df = pd.DataFrame(confusion, index=class_labels, columns=class_labels)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_df, annot=True, fmt="d", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
