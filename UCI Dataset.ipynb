```python
# Calculate the correlation matrix
correlation_matrix = dataset_mat.corr().round(3)

# Filter the correlation matrix based on the condition
high_correlation = correlation_matrix[(correlation_matrix > 0.5) | (correlation_matrix < -0.5)]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(20, 16))

# Create a correlation matrix table
table = ax.table(cellText=high_correlation.values,
                 rowLabels=high_correlation.index,
                 colLabels=high_correlation.columns,
                 loc='center')

# Modify the table appearance
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Remove axis labels
ax.axis('off')

# Display the table
plt.show()
```
